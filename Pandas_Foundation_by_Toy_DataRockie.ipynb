## Merge

```python
left = {
    'key' : [1,2,3,4],
    'name' : ['a','b','v','n'],
    'age' : [3,5,7,7]
}

right = {
    'key' : [1,2,3,4],
    'city' : ['aa','bb','vcc','nc'],
    'zip' : [553,555,755,666]
}

df_left = pd.DataFrame(left)
df_right = pd.DataFrame(right)
```

```python
df_left
```

```python
df_result = pd.merge(df_left, df_right, on = 'key')
df_result
```
